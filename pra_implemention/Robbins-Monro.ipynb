```python
import sympy as sym
from sympy import*
x=Symbol('x')#xを変数として扱う。
a=1995
b=7
c=15

val = 1
loopnum = 100
stepsize  = 0.0001#これをうまく決めなければ収束しない！！大きい数字だと一定になる場所を通り過ぎてしまったり、そこまでいかい可能性がある。
```

```python
#sym.init_printing()
F=-a*x**2+b*x+c

plot(F)
print("関数F:%s" %F)
print("関数Fの微分:%s" %diff(F,x))
print("関数Fの微分を=0として解いた値:%s" %solve(diff(F,x)))
```

```python
def objective(val):
    energy = -a*val**2+b*val+c
    return energy
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

record= np.zeros(loopnum) 
Erecord= np.zeros(loopnum) 

for k in range(loopnum):#loopnumの回数だけ下の処理を繰り返す。
    
    grad = -2*a*val  + b
    update = grad* stepsize
    val = val + update#valを少しずつずらしていく。
    
    record[k] = val
    Erecord[k] = objective(val)#valをどんどんずらしていった時のobjective(val)の値。

idx = np.array(range(loopnum))

plt.figure(figsize = (10,10) )#画像を大きさを調節。
plt.plot(idx, Erecord)
```
