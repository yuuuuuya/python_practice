```python
xmin,xmax=-np.pi,np.pi
x=np.arange(xmin,xmax,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(xmin,xmax) #描写するxの範囲を決める。
plt.ylim(-1,1)
```

```python
# xとyの値は先ほどと共通

# sinのプロット
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title(r"$\sin x$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)

# cosのプロット
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title(r"$\cos x$")
plt.xlim(xmin, xmax)
plt.ylim(-1.3, 1.3)

plt.tight_layout() #タイトルの被りを防ぐ。 これがないと、(‐1，1)の表示がされる。
```

```python
#sinのプロット
plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title("$\sin x$")
plt.xlim(xmin,xmax)
plt.ylim(-1,1)

#cosのプロット
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title("$\cos x$")   #\がないと文字が傾く。　rは何？？  
plt.xlim(xmin,xmax)
plt.ylim(-1,1)

#タイトルがかぶっている。
```

```python
def plot_function_name(name,x=0, y=0):  #x,yは文字の位置
    plt.text(x, y, name, alpha=0.5, size=25, ha="center", va="center") #アルファベットの色の濃さ


x = np.arange(-3, 3.01, 0.01) # 3.01

plt.subplot(2,3,1)
plot_function_name("sin")
plt.plot(x, np.sin(x))

plt.subplot(2,3,2)
plot_function_name("cos")
plt.plot(x, np.cos(x))

plt.subplot(2,3,3)
plot_function_name("tan")
plt.plot(x, np.tan(x))
plt.ylim(-1, 1)  # y軸が無限まで行ってしまうので制限

plt.subplot(2,3,4)
plot_function_name("sinh")
plt.plot(x, np.sinh(x))

plt.subplot(2,3,5)
plot_function_name("cosh",x=0,y=6) #coshのグラフはy=0が下の方にあるため、文字の位置をcoshだけ特別に指定。
plt.plot(x, np.cosh(x))

plt.subplot(2,3,6)
plot_function_name("tanh")
plt.plot(x, np.tanh(x))
```
