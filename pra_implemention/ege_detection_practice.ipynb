```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
#opencv install:'pip install opencv-python' on terminal.
#environment:python3, anaconda4.2(64bit)

import sys
sys.path.append('../pra_dataset')
import os.path
```

```python
''

datapath = '../pra_dataset'
epochIdx = 0
record = []
for i in os.listdir(datapath):
    img = cv2.imread(datapath+'/'+i, 0)#'0' to be gray scale.
    record = record +[img]
    epochIdx += 0

img = record[1]
plt.imshow(img,interpolation  = 'none')
```

```python
aligned_img= img.transpose()
print(aligned_img.shape)
```

```python
print(np.array(range(aligned_img.shape[0])))
```

```python
new_Xindices = aligned_img.shape[0] - np.array(range(aligned_img.shape[0])) -1
print(new_Xindices)
```

```python
plt.imshow(aligned_img[new_Xindices], interpolation = 'none') 
```

```python
img2 = aligned_img[new_Xindices]

laplacian = cv2.Laplacian(img2,cv2.CV_64F)#1img
sobelx = cv2.Sobel(img2,cv2.CV_64F,1,0,ksize=5)#2bit
sobely = cv2.Sobel(img2,cv2.CV_64F,0,1,ksize=5)#3x11

plt.figure(figsize=(20,20))#

plt.subplot(2,2,1),plt.imshow(img2,cmap = 'gray')#plt.subplot(2, 1, 1)221cmap
plt.title('Original'), plt.xticks([]), plt.yticks([])#x,y

#plt.xticksplt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
#plt.yticks([-1, 0, +1])x-pi,-pi/2,0,pi/2,piy-1,1,0


plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray', interpolation='none')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray', interpolation='none')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray', interpolation='none')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```
