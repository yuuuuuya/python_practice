```python
F1 = (x + x**2)/(x*sin(y)**2 + x*cos(y)**2)
F2 = x**2 + y**2 -2
F3 = 2*x**2 -3*x +1
A = Matrix([[1,F2],[2,F3]])
```

```python
B = Matrix([F1,ratsimp(F1),trigsimp(F1),cancel(F1),cancel(trigsimp(F1))])
transpose(A)#もとの関数、簡約化、別の簡約化、xでキャンセル、3つ目要素をxでキャンセル。
```

```python
plot_implicit(F1,(x,-2,2),(y,-2,2))
plot_implicit(F2,(x,-2,2),(y,-2,2))#plot_implicitは2変数関数のplot
plot(F3)
```

```python
C = Matrix([F3,F3.factor(),F3.subs([(x,0)])])
transpose(B)#もとの関数、因数分解、x=0を代入した値。
```

```python
solve(Eq(F3, 0),x)#F3=0となるxのあたい。
```

```python
grad = diff(F3,x)
grad#微分
```

```python
grad.subs([(x,0)])#傾き
```

```python
integrate(F3,x)#積分
```

```python
integrate(F3,(x,-1,3))#定積分
```
