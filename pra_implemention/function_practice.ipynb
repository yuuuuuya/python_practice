```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
from sympy import*
```

```python
x = Symbol('x')
sigmoid = 1/(1+exp(-x))
plot(sigmoid)
```

```python
def dis(x):
    sigmoid = 1/(1+exp(-x))
    return(sigmoid)
```

```python
capsized = np.array([dis(k) for k in [-10,10]])
```

```python
print(capsized)
plt.plot(capsized)
```
